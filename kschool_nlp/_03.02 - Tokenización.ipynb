# NLTK

### Tokenización en frases

###### nltk.RegexpTokenizer

```python
import nltk
```

```python
t = nltk.RegexpTokenizer('\.', gaps = True)
```

```python
t.tokenize(text)
```

###### gaps = True

```python
pattern = r'\.'
```

```python
pattern = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<![A-Z]\.)(?<=\.|\?|\!)\s'
```

```python
t = nltk.RegexpTokenizer(pattern, gaps = True)
```

```python
t.tokenize(text)
```

###### nltk.sent_tokenize

```python
nltk.sent_tokenize(text)
```

```python
quijote = open("./data/quijote.txt", encoding = "utf8").read()
quijote[:600]
```
