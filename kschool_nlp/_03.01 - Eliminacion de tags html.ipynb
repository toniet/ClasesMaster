```python
url = 'https://es.wikipedia.org/wiki/Canis_familiaris'
```

```python
import requests
```

```python
raw_content = requests.get(url).content
```

```python
raw_content
```

```python
import bs4
```

```python
html_content = bs4.BeautifulSoup(raw_content, 'html.parser')
```
