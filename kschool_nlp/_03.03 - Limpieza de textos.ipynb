# Eliminacin de acentos

```python
import unicodedata
```

```python
text = "Brjula"
```

```python
unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf8", "ignore")
```

###### Funcin remove_accents

```python
def remove_accents(text):
    return unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf8", "ignore")
```

```python
remove_accents('ste es mi camin')
```

# Expansin de contracciones

```python
text = "I'm happy"
```

```python
import re
```

```python
re.sub("'m", ' am', text)
```

###### Funcin expand_contractions

```python
def expand_contractions(text):
    text = re.sub("'m", ' am', text)
    text = re.sub("'ve", ' have', text)
    text = re.sub("n't", 'n not', text)
    return text
```

```python
text = "you can't do that!"
```

```python
expand_contractions(text)
```

# Eliminacin de caracteres especiales

```python
pattern = r"[^a-zA-Z0-9\s]"
```

```python
text = "123### What the duck!!!!!"
```

###### Funcin remove_special_characters

```python
def remove_special_characters(text):
    pattern = r"[^a-zA-Z0-9\s]"
    return re.sub(pattern, '', text)
```

```python
text = "123### What???"
```

```python
remove_special_characters(text)
```

# Eliminacin de stopwords

```python
import nltk
```

```python
fileids = ["austen-emma.txt", "austen-persuasion.txt", "austen-sense.txt"]
corpus = ""
for file in fileids:
    corpus += nltk.corpus.gutenberg.raw(file)
```

```python
corpus
```

```python
len(corpus)
```

```python
f = nltk.FreqDist(nltk.word_tokenize(corpus))
```

```python
f
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,4))
f.plot(50)
plt.show()
```

```python
len(nltk.corpus.stopwords.words(fileids='english')), len(nltk.corpus.stopwords.words(fileids='spanish'))
```

###### Funcin remove_stopwords (remove_not)

```python
def remove_stopwords(text):
    words = nltk.word_tokenize(text)
    stopwords = nltk.corpus.stopwords.words(fileids='english')
    return [word for word in words if word not in stopwords]
```

```python
text = "my house is not yours"
```

```python
remove_stopwords(text)
```

# Eliminacin de documentos vacos

```python
corpus = ["house", "", "dog", " "]
```

###### Funcin remove_empty_words

```python
def remove_empty_word(text):
    return [word.strip() for word in text if word.strip() != '']

```

```python
remove_empty_word(corpus)
```
